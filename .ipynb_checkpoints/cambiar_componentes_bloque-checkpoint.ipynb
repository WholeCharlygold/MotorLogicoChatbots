```python
def compiladorPybars(source,jsonData):
    compiler = pybars.Compiler()
    if type(source)==dict:
        source=json.dumps(source)   
    template = compiler.compile(source)
    output = template(jsonData)
    try:
        json_acceptable_string = output.replace("'", "\"")
        datos = json.loads(json_acceptable_string)
        return datos
    except:
        return output
```

```python
def preLlenado(nameJson,datosCargarFuncionalidad,datosFuncionalidadLlenado):
    ''' funcion para simular intruduccion de llenado 
        nameJson: cadena con nombre del json que guarda las funcionalidades con los vales introducidos por el usuario
        datosCargarFuncionalidad: 
        datosFuncionalidadLlenado: diccionario que contiene los datos itroducidos por el usuario 
    '''
    
    for funcionalidad, datosFuncionalidad in datosFuncionalidadLlenado.items():
        print(datosFuncionalidadLlenado[funcionalidad])
        if datosFuncionalidadLlenado[funcionalidad]["type"] == "menuFijo":
            print("visual")
        else: 
            print(datosFuncionalidadLlenado[funcionalidad]["type"])
            

        for componente, datosComponente in datosFuncionalidad["requerimientos"].items():                       
                for keyDato,valueDato in datosComponente.items():
                    #agregamos a los requerimientos la llave value para indicar el dato
                    datosCargarFuncionalidad[funcionalidad]["requerimientos"][componente][keyDato].update({"value":valueDato})
        print("outputs: ",datosFuncionalidad["outputs"])  
        for componente, datosComponente in datosFuncionalidad["outputs"].items():
            
            print("\n\ŧ datos componente",datosComponente)
            for keySalida,valueSalida in datosComponente.items():
                #agregamos a las salidas la llave value para indicar el dato
                datosCargarFuncionalidad[funcionalidad]["outputs"][componente][keySalida].update({"value":valueSalida})      

    print(datosCargarFuncionalidad)
    manejadorArchivos.jsonUnico(nameJson, datosCargarFuncionalidad)
```
